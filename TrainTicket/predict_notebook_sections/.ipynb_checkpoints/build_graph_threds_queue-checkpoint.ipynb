```python
import threading
import queue
import time
import networkx as nx
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import shift
from sklearn.metrics import mean_squared_error, r2_score
from scipy.optimize import curve_fit
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import random
import itertools
```

```python
# -- Functions

def calculate_granger_causality(ts1, ts2):

    # Perform Granger causality test
    data = pd.DataFrame({'ts1': ts1, 'ts2': ts2})
    results = sm.tsa.stattools.grangercausalitytests(data, maxlag=1, verbose=False)

    # Get the F-statistic and p-value from the test
    f_statistic = results[1][0]['ssr_ftest'][0]
    p_value = results[1][0]['ssr_ftest'][1]

    return f_statistic, p_value

# Define a non-linear function (e.g., a quadratic function)
def non_linear_model_one_variable(x, a, b, c):
    return a * x**2 + b * x + c

def non_linear_model_two_variables(x, a, b, c, d):
    # x is a tuple containing two predictors (x1 and x2)
    x1, x2 = x
    return a * x1**2 + b * x2 + c * x1 + d

def calculate_causality_strength(x, y, lag):
    
    """
    x - independent varianble
    y - dependant varianble
    """

    # lagged values of ts_1
    x_lagged = shift(x, lag)
    
    # lagged values of ts_2
    y_lagged = shift(y, lag)
    
    # Fit the non-linear model with one variable to the data
    params, _ = curve_fit(non_linear_model_one_variable, y_lagged, y)

    # Calculate the pseudo-R-squared for the non-linear model with one variable
    SSR = np.sum((y - non_linear_model_one_variable(y_lagged, *params)) ** 2)
    SST = np.sum((y - np.mean(y_lagged)) ** 2)
    pseudo_R_squared_r = 1 - (SSR / SST)

    # Fit the non-linear model with two variables to the data
    params, _ = curve_fit(non_linear_model_two_variables, (y_lagged, x_lagged), y)

    # Calculate the pseudo-R-squared for the non-linear model with two variables
    SSR = np.sum((y - non_linear_model_two_variables((y_lagged, x_lagged), *params)) ** 2)
    SST = np.sum((y - np.mean(y_lagged)) ** 2)
    pseudo_R_squared_unr = 1 - (SSR / SST)
    
    # Compare the two pseudo-R-squared values
    # print(f"Pseudo-R-squared of the restricted model: {pseudo_R_squared_r:.4f}")
    # print(f"Pseudo-R-squared of the unrestricted model: {pseudo_R_squared_unr:.4f}")

    # Compare the improvement in R-squared
    causality_strength = round(pseudo_R_squared_unr - pseudo_R_squared_r, 4)
    # print(f"Improvement in R-squared: {causality_strength:.4f}")
    
    if causality_strength < 0:
        causality_strength = 0

    return causality_strength


# Visualize the graph
def graph_visualize(G):

    # Position nodes using a spring layout
    pos = nx.spring_layout(G, seed=42)
    labels = {node: node for node in G.nodes()}

    plt.figure(figsize=(10, 10))
    nx.draw(G, pos, with_labels=True, labels=labels, node_size=800, node_color='lightblue', font_size=12, font_color='black', font_weight='bold', arrowsize=20)
    plt.title("Causal Dependency Graph")
    plt.show()


# Define a function that represents a thread fetching data from a pull source
def data_consumer(thread_id, data_queue):
    while True:
        try:
            item = data_queue.get(timeout=1)  # Get data from the queue (with timeout)
        except queue.Empty:
            print(f"Thread {thread_id} finished.")
            break
        else:
            # print(f"Thread {thread_id} got data: {item}")
            
            
            kpi_1, series_1 = item[0][0], item[0][1]
            kpi_2, series_2 = item[1][0], item[1][1]
            
            print(f"Thread {thread_id} got data: {kpi_1} {kpi_2}")
            
            f_stat, p_val = calculate_granger_causality(series_1, series_2)
            if p_val < significance_level:

                causality_strength = calculate_causality_strength(series_1, series_2, lag)
                    
                print(kpi_1, kpi_2, causality_strength)
                
                if not G.has_edge(kpi_1, kpi_2):
                    
                    add_node(G, kpi_1)
                    add_node(G, kpi_2)
                    G.add_edge(kpi_2, kpi_1, weight=causality_strength)
                else:
                    edge_weight = G[kpi_1][kpi_2].get('weight', None)
                    if (edge_weight is None) or (edge_weight < causality_strength):
                        G.add_edge(kpi_2, kpi_1, weight=causality_strength)

            data_queue.task_done()
            
def add_node(graph, node_name):
    node_exists = False
    for existed_node in graph.nodes(data=True):
        if node_name == existed_node[0]:
            node_exists = True
            break

    if not node_exists:
        graph.add_node(node_name)
```

```python
# -- Generate granger caused data

n = 20

kpis = {}
for ii in range(1, 3):

    # Generate synthetic time series data
    np.random.seed(0)

    # Generate a time series for X
    t = np.arange(n)
    X = 0.5 * t + np.random.normal(scale=2, size=n)

    # Generate a time series for Y that is influenced by X (Granger-caused by X)
    Y = 0.7 * X[:-1] + np.random.normal(scale=1, size=n-1)
    Y = np.concatenate(([0], Y))  # Make Y the same length as X
    
    kpis["k_" + str(ii)] = X
    kpis["k_" + str(ii) + "1"] = Y


    # Plot the time series data
    plt.figure(figsize=(10, 4))
    plt.plot(t, X, label='X')
    plt.plot(t, Y, label='Y')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.legend()
    plt.title('Synthetic Time Series Data')
    plt.grid(True)
    plt.show()
```

```python
# -- Check granger causality and calculate the r-square

lag = 1
significance_level = 0.05
# Create a shared queue to store data
data_queue = queue.Queue()

# -- Build the graph

global G
G = nx.DiGraph()

# Add data to the queue (you can replace this with your data source)
for pair in itertools.permutations(kpis, 2):
    kpi_1, kpi_2 = pair
    
    pair = ((kpi_1, kpis[kpi_1]), (kpi_2, kpis[kpi_2]))
    
    data_queue.put(pair)

# Create and start multiple threads
num_threads = 2
threads = []

for i in range(num_threads):
    thread = threading.Thread(target=data_consumer, args=(i, data_queue))
    thread.start()
    threads.append(thread)

# Wait for all threads to finish
for thread in threads:
    thread.join()

print("\nAll threads have finished.\n")
```

```python
# -- Graph data

graph_visualize(G)

print(G)

print("\nEdge Weights:")
for u, v, data in G.edges(data=True):
    if 'weight' in data:
        print(f"({u}, {v}) : {data['weight']}")
```
